```python
df = an2.copy()
```

```python
cur = df.currency.unique()
cur
```

```python
def get_cur(cur):
    result = requests.get(f'https://open.er-api.com/v6/latest/USD')
    data = result.json()
    rat = data['rates']
    dict_cur = {}
    for c in cur:
        if c in rat:
            result = requests.get(f'https://open.er-api.com/v6/latest/{c}')
            data = result.json()
            r = data['rates']['RUB']
            dict_cur[c] = r
    return dict_cur
```

```python
rates = get_cur(cur)
rates['RUR'] = 1
rates
```

```python
# rates = dict(RUR=1, USD=90, EUR=100)
```

```python
df['from_'] = df.currency.map(rates) * df.s_from
```

```python
df['to_'] = df.currency.map(rates) * df.s_to
```

```python
import scipy
import scipy.stats as stats
stats.shapiro(df.from_)
```

```python
# коэффициенты ассимметрии и эксцесса
skew = df.from_.skew()
kurt = df.from_.kurt()
print('Коэффициент ассиметрии:', skew)
print('Коэффициент эксцесса:', kurt)
```
